```python
#时序数据滑窗转换用于将时间序列数据转为回归数据，简单地说，就是把一个单序列的数据变为X->Y的回归数据。

#步阶为2代表2个X（步阶多少就有多少个X），一个Y（这个不会变的），

#简单地说，就是用第1，2天的数据预测第3天，用第2，3天的数据预测第4天，以此类推。
def create_dataset(dataset, look_back):

    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back)]
        dataX.append(a)
        dataY.append(dataset[i + look_back])
        dddd=pd.concat([pd.DataFrame(np.array(dataY)),pd.DataFrame(np.array(dataX))],axis=1)
        dddd.columns=['Y']+['shiftX_'+str(i) for i in range(len(dddd.columns)-1)]
    return dddd



    
data1=create_dataset(temp3['批发价格(元/千克)'],look_back=1)
data1.dropna(inplace=True)
data1
```

```python
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.metrics import mean_squared_error #均方误差
from sklearn.metrics import mean_absolute_error #平方绝对误差

def score(y_true, y_pre):
    # MSE
    print("MSE :")
    print(metrics.mean_squared_error(y_true, y_pre)) 
    # RMSE
    print("RMSE :")
    print(np.sqrt(metrics.mean_squared_error(y_true, y_pre))) 
    # MAE
    print("MAE :")
    print(metrics.mean_absolute_error(y_true, y_pre)) 
    # R2
    print("R2 :")
    print(r2_score(y_true,y_pre)) 
```

```python
'''
我们需要编写一个规划求解模型，并使用遗产算法进行求解，其中：
Wholesale=8.411665843100959#批发价
Loss_rate=8.758815129904537#损耗率
x#销售单价
Sales_unit_price=xgb_model.predict(np.array([x,Wholesale]).reshape(1,-1))#销售数量

目标函数为：max (Sales_unit_price*(100-Loss_rate)/100)*(x-Wholesale)

约束条件为：x>Wholesale

在上面的基础上，我们知道未来一周的批发价为[8.411665843100959,
 8.960400386043592,
 9.215648220289571,
 8.971439743052331,
 7.876682258094114,
 7.42452926566904,
 7.043501862916413]（含第一天），损耗率均为.758815129904537，请给出并给出各蔬菜品类未来一周的销售量和销售单价， 使得商超收益最大

'''
```

```python
import numpy as np
from deap import base, creator, tools, algorithms
import xgboost as xgb
Wholesale=8.411665843100959#批发价
Loss_rate=8.758815129904537#损耗率
x=Wholesale*temp3['平均每千克差价占比批发价格比例'].max()
Sales_unit_price=xgb_model.predict(np.array([x,Wholesale]).reshape(1,-1))#销售数量

def predict_sales_unit_price(selling_price, wholesale_price):
    return xgb_model.predict(np.array([selling_price, wholesale_price]).reshape(1, -1))[0]

def objective(individual, wholesale_price):
    selling_price = individual[0]
    Sales_unit_price = predict_sales_unit_price(selling_price, wholesale_price)
    profit = (Sales_unit_price * (100 - Loss_rate) / 100) * (selling_price - wholesale_price)
    return profit,

def feasible(individual, wholesale_price):
    return individual[0] > wholesale_price

def optimize_price_for_day(wholesale_price):
    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("attr_float", np.random.uniform, wholesale_price, wholesale_price*1.5)
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=1)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("mate", tools.cxBlend, alpha=0.5)
    toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.1)
    toolbox.register("select", tools.selTournament, tournsize=3)
    
    toolbox.register("evaluate", objective, wholesale_price=wholesale_price)
    toolbox.decorate("evaluate", tools.DeltaPenalty(lambda ind: feasible(ind, wholesale_price), -np.inf))

    pop = toolbox.population(n=10)
    algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=40, verbose=False)

    best_individual = tools.selBest(pop, k=1)[0]
    return best_individual[0]

wholesale_prices = [
    8.411665843100959, 8.960400386043592, 9.215648220289571, 
    8.971439743052331, 7.876682258094114, 7.42452926566904, 7.043501862916413
]
Loss_rate = 8.758815129904537

optimal_prices = []
sales_volumes = []
profits = []
for wholesale_price in wholesale_prices:
    optimal_selling_price = optimize_price_for_day(wholesale_price)
    optimal_sales_volume = predict_sales_unit_price(optimal_selling_price, wholesale_price)
        # Calculate profit for this day
    day_profit = (optimal_sales_volume * (100 - Loss_rate) / 100) * (optimal_selling_price - wholesale_price)
    profits.append(day_profit)
    optimal_prices.append(optimal_selling_price)
    sales_volumes.append(optimal_sales_volume)

print("Optimal Selling Prices:", optimal_prices)
print("Sales Volumes:", sales_volumes)
print("Daily Profits:", profits)
print("Total Profit:", sum(profits))
```

```python
import numpy as np
import xgboost as xgb
from pyswarm import pso
Wholesale=8.411665843100959#批发价
Loss_rate=8.758815129904537#损耗率
x=Wholesale*temp3['平均每千克差价占比批发价格比例'].max()
Sales_unit_price=xgb_model.predict(np.array([x,Wholesale]).reshape(1,-1))#销售数量




def objective_pso(selling_price, *args):
    wholesale_price = args[0]
    Sales_unit_price = predict_sales_unit_price(selling_price, wholesale_price)
    profit = -(Sales_unit_price * (100 - Loss_rate) / 100) * (selling_price - wholesale_price)  # 注意：我们用-号来转换问题为最小化
    return profit

def constraint(selling_price, *args):
    wholesale_price = args[0]
    return selling_price - wholesale_price

def optimize_price_for_day_pso(wholesale_price):
    lb = [wholesale_price]  # lower bound
    ub = [wholesale_price * 1.5]  # upper bound for example

    # 使用PSO进行优化
    xopt, fopt = pso(objective_pso, lb, ub, f_ieqcons=constraint, args=(wholesale_price,))
    return xopt[0]

wholesale_prices = [
    8.411665843100959, 8.960400386043592, 9.215648220289571, 
    8.971439743052331, 7.876682258094114, 7.42452926566904, 7.043501862916413
]
Loss_rate = 8.758815129904537

optimal_prices_pso = []
sales_volumes_pso = []

for wholesale_price in wholesale_prices:
    optimal_selling_price_pso = optimize_price_for_day_pso(wholesale_price)
    optimal_sales_volume_pso = predict_sales_unit_price(optimal_selling_price_pso, wholesale_price)
    optimal_prices_pso.append(optimal_selling_price_pso)
    sales_volumes_pso.append(optimal_sales_volume_pso)

print("Optimal Selling Prices using PSO:", optimal_prices_pso)
print("Sales Volumes using PSO:", sales_volumes_pso)


```
